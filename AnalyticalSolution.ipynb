### Analytical Solution

For the CARA Utility function:
$$ 
U(W_{T})=\frac{1-e^{-\alpha W_{T}}}{\alpha}, \alpha\neq 0
$$
Since we as assume no consumption of wealth from $t = 0,1,...,T-1$
We can smplified the MDP reward at time $T$ as
$$
\frac{-e^{-\alpha W_{T}}}{\alpha}
$$
So we can denote the Value function at time t given policy $\pi$ as:
$$
V^{\pi}_{t}(W_{t})=\mathbb{E}_{\pi}[\frac{-e^{-\alpha W_{T}}}{\alpha}|(t,W_{t})]
$$
and the Bellman Optimality Equation is:
$$
V^{*}_{t}(W_{t}) = \max_{{x_t}}Q^{*}_{t}(W_{t},x_t)=\max_{{x_t}}\mathbb{E}_{Y_t}[V^{*}_{t+1}(W_{t+1})] 
$$
for $t=0,1,...,T-2$, and
$$
V^{*}_{T-1}(W_{T-1}) = \max_{{x_{T-1}}}Q^{*}_{T-1}(W_{T-1},x_{T-1})=\max_{{{x_{T-1}}}}\mathbb{E}_{Y_{T-1}}[\frac{-e^{-\alpha W_{T}}}{\alpha}] 
$$
for $t=T-1$
Using an educated guess of the functional form of Optimal Value Function:
$$
V^{*}_{t}(W_{t})=-b_te^{c_tW_t}
$$

So, the Bellman Optimality Equation can be writtten as
$$V^{*}_{t}(W_{t})=\max_{{x_t}}\mathbb{E}_{Y_t}[-b_{t+1}e^{c_{t+1}W_{t+1}}]\\
=\max_{{x_t}}\{\mathbb{E}_{Y_{t}}[-b_{t+1}e^{-c_{t+1}((Y_{t}-r)x_{t}+(1+r)W_{t})}]\}\\
=\max_{{x_t}}\{-b_{t+1}e^{-c_{t+1}(1+r)W_{t}}\mathbb{E}_{Y_{t}}[e^{-c_{t+1}(Y_{t}-r)x_{t}}]\}\\
=\max_{{x_t}}\{-b_{t+1}e^{-c_{t+1}(1+r)W_{t}}[pe^{-c_{t+1}(a-r)x_{t}}+(1-p)e^{-c_{t+1}(b-r)x_{t}}]\}\\
\Rightarrow Q^{*}_{t}(W_{t},x_{t})=-b_{t+1}e^{-c_{t+1}(1+r)W_{t}}[pe^{-c_{t+1}(a-r)x_{t}}+(1-p)e^{-c_{t+1}(b-r)x_{t}}]\\
arg\max_{{x_t}}Q^{*}_{t}(W_{t},x_{t}) \Longleftrightarrow \frac{\partial}{\partial x_t}[pe^{-c_{t+1}(a-r)x_{t}}+(1-p)e^{-c_{t+1}(b-r)x_{t}}]=0\\
\Rightarrow p(-c_{t+1})(a-r)e^{-c_{t+1}(a-r)x_{t}}+(1-p)(-c_{t+1})(b-r)e^{-c_{t+1}(b-r)x_{t}}=0\\
\Rightarrow p(a-r)e^{-c_{t+1}(a-r)x_{t}}=(p-1)(b-r)e^{-c_{t+1}(b-r)x_{t}}\\
\Rightarrow e^{c_{t+1}(a-r)x_{t}-c_{t+1}(b-r)x_{t}}=\frac{p(a-r)}{(p-1)(b-r)}\\
\Rightarrow c_{t+1}x_t(a-b)=ln\frac{p(a-r)}{(p-1)(b-r)}\\
\Rightarrow x^*_t=\frac{ln\frac{p(a-r)}{(p-1)(b-r)}}{c_{t+1}(a-b)}
$$
Note that $$p\in(0,1)$$
and We require 
$$
\frac{p(a-r)}{(p-1)(b-r)}>0 \Rightarrow \frac{a-r}{b-r}<0\\
\Rightarrow r \in(b,a)
$$
Also note that
$$
\frac{\partial^2}{\partial x^2_t}Q^{*}_{t}(W_{t})\\
=-b_{t+1}e^{-c_{t+1}(1+r)W_{t}}[p(c^2_{t+1})(a-r)^2e^{c_{t+1}(a-r)x_{t}}+(1-p)(c^2_{t+1})(b-r)^2e^{-c_{t+1}(b-r)x_{t}}]<0
$$
So, $x_t$ is ensured to achieve the global maximum

for $t=T-1$, we maximize the Reward function $\frac{-e^{-\alpha W_{T}}}{\alpha}$, so,
$$
V^{*}_{T-1}(W_{T-1})=\max_{{x_{T-1}}}\{\mathbb{E}_{Y_t}[\frac{-e^{-\alpha W_{T}}}{\alpha}]\}\\
=\max_{{x_{T-1}}}\{\mathbb{E}_{Y_t}[\frac{-e^{-\alpha((Y_{t}-r)x_{t}+(1+r)W_{t})}}{\alpha}]\}\\
=\max_{{x_{T-1}}}\{\frac{-e^{-\alpha(1+r)W_{t}}}{\alpha}\mathbb{E}_{Y_t}[-e^{-\alpha (Y_{t}-r)x_{t}}]\}
$$
Treating $\alpha = c_{t+1}=c_T$, we have,
$$
x^*_{T-1}=\frac{ln\frac{p(a-r)}{(p-1)(b-r)}}{\alpha(a-b)}
$$

substitute $x^*_{T-1}$ into $V^{*}_{T-1}(W_{T-1})$
$$
V^{*}_{T-1}(W_{T-1})=\frac{-1}{\alpha}[pe^{(r-a)\frac{ln\frac{p(a-r)}{(p-1)(b-r)}}{(a-b)}}+(1-p)e^{(r-b)\frac{ln\frac{p(a-r)}{(p-1)(b-r)}}{(a-b)}}]e^{-\alpha(1+r)W_{T-1}}\\
\Rightarrow \begin{cases}
  b_{T-1} = \frac{1}{\alpha}[pe^{(r-a)\frac{ln\frac{p(a-r)}{(p-1)(b-r)}}{(a-b)}}+(1-p)e^{(r-b)\frac{ln\frac{p(a-r)}{(p-1)(b-r)}}{(a-b)}}]\\
  c_{T-1} = \alpha(1+r)    
\end{cases}
$$

for general $t={0,1,...,T-2}$
$$
V^{*}_{t}(W_{t})=-b_{t+1}[pe^{(r-a)\frac{ln\frac{p(a-r)}{(p-1)(b-r)}}{(a-b)}}+(1-p)e^{(r-b)\frac{ln\frac{p(a-r)}{(p-1)(b-r)}}{(a-b)}}]e^{-c_{t+1}(1+r)W_{t}}
$$
So,
$$
\Rightarrow b_{T-2}=b_{T-1}[pe^{(r-a)\frac{ln\frac{p(a-r)}{(p-1)(b-r)}}{(a-b)}}+(1-p)e^{(r-b)\frac{ln\frac{p(a-r)}{(p-1)(b-r)}}{(a-b)}}]\\
=\frac{1}{\alpha}[pe^{(r-a)\frac{ln\frac{p(a-r)}{(p-1)(b-r)}}{(a-b)}}+(1-p)e^{(r-b)\frac{ln\frac{p(a-r)}{(p-1)(b-r)}}{(a-b)}}]^2\\
...\\
\Rightarrow b_{t} = \frac{1}{\alpha}[pe^{(r-a)\frac{ln\frac{p(a-r)}{(p-1)(b-r)}}{(a-b)}}+(1-p)e^{(r-b)\frac{ln\frac{p(a-r)}{(p-1)(b-r)}}{(a-b)}}]^{(T-t)}\\
$$
and
$$
\Rightarrow c_{T-2}=c_{T-1}(1+r)\\
\Rightarrow c_{T-2}=\alpha(1+r)^2\\
...\\
\Rightarrow c_{t}=\alpha(1+r)^{T-t}
$$

$$
\Rightarrow \begin{cases}
  x^*_{t} = \frac{ln\frac{p(a-r)}{(p-1)(b-r)}}{\alpha(1+r)^{(T-t-1)}(a-b)}\\
  b_{t} = \frac{1}{\alpha}[pe^{(r-a)\frac{ln\frac{p(a-r)}{(p-1)(b-r)}}{(a-b)}}+(1-p)e^{(r-b)\frac{ln\frac{p(a-r)}{(p-1)(b-r)}}{(a-b)}}]^{(T-t)}\\
  c_{t}=\alpha(1+r)^{T-t} 
\end{cases}
$$

### Code

```python
import numpy as np
from scipy.stats import bernoulli, norm
import random
import matplotlib.pyplot as plt
%matplotlib inline
from tqdm.notebook import tnrange
```

```python
def Utility_CARA(WT,alpha):
    return (1-np.exp(-alpha*WT))/alpha

def b_multiplier(a,b,p,r):
    ln_fun = np.log((p*(a-r))/((p-1)*(b-r)))
    m1 = p*np.exp((r-a)*ln_fun/(a-b))
    m2 = (1-p)*np.exp((r-b)*ln_fun/(a-b))
    return(m1+m2)

def bt(a,b,p,r,T,t,alpha):
    bmul = b_multiplier(a,b,p,r)
    return(bmul**(T-t)/alpha)

def ct(alpha,r,T,t):
    return(alpha*(1+r)**(T-t))

def xt(a,b,p,r,t,alpha):
    return(np.log((p*(a-r))/((p-1)*(b-r)))/(alpha*(1+r)**(T-t-1)*(a-b)))
    
def Qt(bt1,ct1,Wt,xt):
    C = -bt1*np.exp(-ct1*(1+r)*Wt)
    Ep = p*np.exp(-ct1*(a-r)*xt)
    Eq = (1-p)*np.exp(-ct1*(b-r)*xt)
    return(C*(Ep+Eq))

def Vt(bt,ct,Wt):
    return(-bt*np.exp(-ct*Wt))

def Yt(a,b,p,n=1):
    return bernoulli.rvs(size = n,p = p)*(a-b)+b

def Wt(W0,xt,r,Yt,a,b,p):
    t = len(xt)
    W = np.zeros(t+1)
    W[0] = W0
    for i in range(t):
        W[i+1] = (W[i]-xt[i])*(1+r) + xt[i]*(1+Yt(a,b,p))
    return W
```

```python
a = 0.08
b = 0.02
p = 0.6
r = 0.03
T = 10
paths = 10000
alpha = 1
```

```python
def printPath(path,alpha,t,T,bt1_10,ct1_10,xt0_9,W0_10):
    print("Sample Path " + str(path))
    for t in range(T):
        print("----------------------------------------------------")
        print("Period " + str(t))
        print("b_"+str(t+1)+" = " + str(bt1_10[t]))
        print("c_"+str(t+1)+" = " + str(ct1_10[t]))
        print("Optimal allocation x_"+str(t)+" = " + str(xt0_9[t]))
        print("Wealth = " + str(W0_10[t,path-1]))
        print("Q function value = " + str(Qt(bt1_10[t],ct1_10[t],W0_10[t,path-1],xt0_9[t])))  
    print("----------------------------------------------------")
    print("Period " + str(T))
    print("Terminal Wealth = " + str(W0_10[T,path-1]))
    print("Utility = " + str(Utility_CARA(W0_10[T,path-1],alpha)))

def Allocation(a,b,p,r,T,alpha,paths):
    t = np.array(range(0,T+1))
    bt0_9 = bt(a,b,p,r,T,t[:-1],alpha)
    bt1_10 = bt(a,b,p,r,T,t[1:],alpha)
    ct0_9 = ct(alpha,r,T,t[:-1])
    ct1_10 = ct(alpha,r,T,t[1:])
    xt0_9 = xt(a,b,p,r,t[:-1],alpha)
    print("xt", xt0_9)
    W0_10 = np.zeros((T+1)*paths).reshape((T+1,paths))
    for _ in tnrange(0,paths):
        W0_10[:,_] = Wt(norm.rvs(loc = 1,scale = 0.1),xt0_9,r,Yt,a,b,p)
    
    path = random.randint(1, paths+1)
    printPath(path,alpha,t,T,bt1_10,ct1_10,xt0_9,W0_10)
    plt.clf()
    plt.plot(W0_10)
    plt.title("risk aversion parameter: " + str(alpha))
    plt.show()
    plt.clf()

    plt.hist(W0_10[T,:],bins = int(paths/10))
    plt.title("Final Wealth Distribution")
    plt.show()
```
